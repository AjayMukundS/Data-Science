```python
#Importing reqired Libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import copy
import category_encoders as ce
%matplotlib inline
```

```python
#Importing the dataset
dataset= pd.read_csv('Data.csv')
X= dataset.iloc[:, :-1].values
y= dataset.iloc[:, 3].values
dataset
```
